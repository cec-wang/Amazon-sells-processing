### Find the trend between ranking in a sub-category to a bigger category.

1. one element polynomial 
<img src="images/single element poly 2.png" />

```python
def one_ele_poly(x, beta_1, beta_2):
    y = beta_1 * np.power(x, beta_2)
    return y
```

```python
Jewelry_x = np.arange(1, Jewelry.shape[0]+1)
Jewelry_BSR = Jewelry.BSR.values
```

```python
Jewelry_to_master = amz.model_fitting(Jewelry_x, Jewelry_BSR, model = one_ele_poly, legend = 'Jewelry')
Jewelry_to_master
```

```python
Earrings_x = np.arange(1, Earrings.shape[0]+1)
Earrings_BSR = Earrings.BSR.values
Earrings_to_master = amz.model_fitting(Earrings_x, Earrings_BSR, one_ele_poly, legend = 'Earrings BSR vs. Sales')
Earrings_to_master
```

```python
beta_1 = Jewelry_to_master['beta_1']
beta_2 = Jewelry_to_master['beta_2']
Earrings_in_Jewelry_ranking = np.power((Earrings_BSR / beta_1), (1/beta_2))
```

```python
Earrings_to_Jewelry = amz.model_fitting(Earrings_x, Earrings_in_Jewelry_ranking, one_ele_poly, legend  = 'Earrings rankings in Jewelry')
Earrings_to_Jewelry
```

2. Exponential 
<img src="images/exp.png" />

```python
def exp(x, beta_1, beta_2, beta_3):
    y = beta_1 * np.exp((x + beta_3) * beta_2)
    return y
```

```python
Asymmetrical_to_master_exp =  amz.model_fitting(Asymmetrical_x, Asymmetrical_BSR, exp, legend = 'Asymmetrical Earrings fit by exponential')
Asymmetrical_to_master_exp
```

```python
popt, pcov = curve_fit(exp, Earrings_x, Earrings_BSR)
```

3. Lorentz Model between the rankings

```python
def Lorentz_ranking(x, beta_1, beta_2):
    y = np.sqrt(beta_1 + beta_2 * np.power(x, 2))
    return y
```

```python
popt, pcov = curve_fit(Lorentz_ranking, Asymmetrical_x, Asymmetrical_BSR, maxfev = 1000)
popt
```

```python
Lorentz_rank = amz.Rank_Fit_Param_Table_generate(Earrings_lst, glob, Lorentz_ranking)
```

Impossible to fit with exponential. Since the derivative of such function is concaving up. However, in our scinerio, sometimes the growth rate is increasing, and sometimes the growth rate is decreasing. 

Same argument can be used for log(x) function as well. According to the derivative, the higher the x and the lower the derivative. So the growth rate is always decreasing. 

The function cannot be polynomial as well. Since I tend to believe the correlation shoud be a smooth line instead of multiple up and downs in the growth rate. Also fitting the curve with multiple component function will increase the complication and making it difficult for future analysis. 

**Conclusion:** The best model for ranking relationship fitting is single element polynomial curve.

1. one element polynomial 
<img src="images/inv single element poly.png" />

```python
def inv_one_ele_poly(x, beta_1, beta_2):
    y = beta_1 / np.power(x, beta_2)
    return y
```

```python
plt.scatter(BSR,Sales)
```

```python
beta_1 = 10000000
beta_2 = 1.00

#logistic function
Sales_pred = inv_one_ele_poly(BSR, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.loglog(BSR,Sales_pred)
plt.loglog(BSR, Sales, 'ro')
print("Mean absolute error: %.2f" % np.mean(np.absolute(Sales-Sales_pred)))
print("Residual sum of squares (MSE): %.2f" % np.mean((Sales-Sales_pred) ** 2))
print("R2-score: %.2f" % r2_score(Sales, Sales_pred))
```
